```python
import json
import whisper

# 1. suppress.json 불러오기
with open("suppress.json", "r") as f:
    suppress_list = json.load(f)

# 2. whisper 토크나이저 불러오기 (multilingual)
tok = whisper.tokenizer.get_tokenizer(multilingual=True, task="transcribe")

# 3. 각 토큰 인덱스가 무슨 문자/기호인지 역으로 변환해서 출력
#   (모든 1~4바이트 utf-8 문자에서 찾기, 효율 위해 영어/기호 우선)
candidates = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ.,?!%@#&*()[]{}<>:;\"'`~^_-+=|\\/0123456789"

# 사전: 토큰 인덱스 → 문자
token_to_char = {}

for ch in candidates:
    token = tok.encode(ch)
    if len(token) == 1:  # 단일 토큰일 때만
        token_to_char[token[0]] = ch

print("=== suppress.json 내역 ===")
for t in suppress_list:
    if t in token_to_char:
        print(f"토큰 {t} : '{token_to_char[t]}' (영어/기호)")
    else:
        print(f"토큰 {t} : (정확한 문자 알 수 없음, 내부 토크나이저 특수토큰일 수도 있음)")
```
